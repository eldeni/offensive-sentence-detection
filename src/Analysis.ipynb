```python
import pandas as pd
import numpy as np
```

```python
trainFile = '../Dataset/quora-insincere-questions-classification/train.csv'
testFile = '../Dataset/quora-insincere-questions-classification/test.csv'
```

```python
train = pd.read_csv(trainFile)
test = pd.read_csv(testFile)
```

```python
train['wordCount'] = train['question_text'].apply(lambda x: len(x.split()))
```

```python
train['characterCount'] = train['question_text'].apply(lambda x: len(x))
```

```python
train.head()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['stopwords'] = train['question_text'].apply(lambda x: len([x for x in x.split() if x in stop]))
train.head()
```

```python
avgWords = train['wordCount'].mean()
avgChars = train['characterCount'].mean()
avgWordsSincere = train.loc[train['target'] == 0]['wordCount'].mean()
avgWordsInsincere = train.loc[train['target'] == 1]['wordCount'].mean()
avgCharsSincere = train.loc[train['target'] == 0]['characterCount'].mean()
avgCharsInsincere = train.loc[train['target'] == 1]['characterCount'].mean()
print(avgWords, avgChars, avgWordsSincere, avgWordsInsincere, \
      avgCharsSincere, avgWordsInsincere)
```
