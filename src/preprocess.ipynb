```python
import nltk
import pandas as pd
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords 
import string
from nltk.tokenize import word_tokenize
```

```python
trainFilePath = '../data/train_sample.csv'
```

```python
df_train = pd.read_csv(trainFilePath)
```

```python
englishStopwords = stopwords.words('english')
print(englishStopwords)
```

```python
def extractNonStopwords(tokenizedSentence):
    return [w for w in tokenizedSentence if not w in englishStopwords]

def stemmize(tokenizedSentence):
    porter = nltk.PorterStemmer()
    stemmedTokens = [porter.stem(t.lower()) for t in tokenizedSentence]
    return stemmedTokens

# https://machinelearningmastery.com/clean-text-machine-learning-python/
def removePunctuation(tokenizedSentence):
    table = str.maketrans('', '', string.punctuation)
    stripped = [token.translate(table) for token in tokenizedSentence]
    return stripped
    
for index, row in df_train.iterrows():
    sentence = row['question_text']
#     print(sentence)
    tokenizedSentence = word_tokenize(sentence)
    sentenceWithoutPunctuation = removePunctuation(tokenizedSentence)
    sentenceWithoutStopwords = extractNonStopwords(sentenceWithoutPunctuation)
    sentenceAfterStemizing = stemmize(sentenceWithoutStopwords)

    print(sentenceAfterStemizing)
```
